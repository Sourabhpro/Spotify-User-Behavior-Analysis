**About the Data:**

**Age -** Age range of users.

**Gender -** Gender of users.

**spotify_usage_period -** Since how long the user is using Spotify.

**spotify_listening_device -** Which devices the user is using.

**spotify_subscription_plan -** Which subscription plan are they having.

**Premium_sub_willingness -** Are they willing to continue or take premium plan

**preffered_premium_plan -** preferred premium plan to take.

**preferred_listening_content -** What type of content they prefer to listen(podcast/music/etc).

**fav_music_genre -** What type of song they prefer to listen.

**music_time_slot -** At what time they prefer listening music.

**music_Influencial_mood -** Music is influenced with what type of mood.
**music_lis_frequency -** At what time user are listening music (ex- leisture time/office time/workout time).

**music_expl_method -** with which method user is listening music.

**music_recc_rating -** music recommendation rating.

**pod_lis_frequency -** Podcast listening frequency.

**fav_pod_genre -** What type of podcast they like.

**preffered_pod_format -** Preferred podcast format like interview, story telling, etc.

**pod_host_preference -** podcast host preference.

**preffered_pod_duration -** Do they prefer shorter podcast episodes (under 30 minutes) or longer episodes (over 30 minutes)

**pod_variety_satisfaction -** Satisfied with the variety and availability of podcasts on Spotify?
